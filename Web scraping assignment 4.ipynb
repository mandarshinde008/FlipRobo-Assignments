```python
##Getting the required theguardian webpage
url = 'https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare'
driver.get(url)
```

```python
detail=[]

details=driver.find_elements_by_xpath('//tbody//tr//td')

for i in details:
    try:
        detail.append(i.text)
    except NoSuchElementException:
        detail.append('--')
detail[:]
```

```python
book=[]
author=[]
volumes=[]
publisher=[]
genre=[]

for i in range(1,len(detail),6):
    book.append(detail[i])
    
for i in range(2,len(detail),6):
    author.append(detail[i])
    
for i in range(3,len(detail),6):
    volumes.append(detail[i])
    
for i in range(4,len(detail),6):
    publisher.append(detail[i])
    
for i in range(5,len(detail),6):
    genre.append(detail[i])
    
best_selling_books=pd.DataFrame()
best_selling_books['Book Name']=book
best_selling_books['Author Name']=author
best_selling_books['Volumes Sold']=volumes
best_selling_books['Publisher']=publisher
best_selling_books['genre']=genre

best_selling_books
```
